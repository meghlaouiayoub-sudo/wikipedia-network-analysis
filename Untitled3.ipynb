```python
edges = pd.read_csv("data/wikipedia/chameleon/musae_chameleon_edges.csv")

edges.head()
```

```python
G = nx.from_pandas_edgelist(edges, source='id1', target='id2')

print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
```

```python
print("Network density:", nx.density(G))
print("Average clustering coefficient:", nx.average_clustering(G))
```

```python
components = list(nx.connected_components(G))

print("Number of components:", len(components))

largest_component = max(components, key=len)

print("Largest component size:", len(largest_component))
```

```python
G_largest = G.subgraph(largest_component)

print("Average path length:", nx.average_shortest_path_length(G_largest))
```

```python
degrees = [d for n, d in G.degree()]

plt.hist(degrees, bins=50)

plt.xlabel("Degree")
plt.ylabel("Number of Nodes")
plt.title("Degree Distribution")

plt.show()
```

```python
centrality = nx.degree_centrality(G)

top_nodes = sorted(centrality.items(), key=lambda x: x[1], reverse=True)[:10]

print("Top nodes by centrality:")
print(top_nodes)
```

```python
n = G.number_of_nodes()
m = G.number_of_edges()

p = (2*m)/(n*(n-1))

ER = nx.erdos_renyi_graph(n, p)
BA = nx.barabasi_albert_graph(n, 5)
WS = nx.watts_strogatz_graph(n, 6, 0.1)
```

```python
print("Original:", nx.average_clustering(G))
print("ER:", nx.average_clustering(ER))
print("BA:", nx.average_clustering(BA))
print("WS:", nx.average_clustering(WS))
```
